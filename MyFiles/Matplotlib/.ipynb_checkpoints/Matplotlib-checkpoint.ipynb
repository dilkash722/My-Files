```python
import matplotlib.pyplot as plt

# Data
products = ['Product A', 'Product B', 'Product C', 'Product D']
sizes = [15, 30, 45, 10]
details = [
    'Detail A: Lorem ipsum dolor sit amet, consectetur adipiscing elit.',
    'Detail B: Nulla facilisi. Proin at nisi sed purus tristique lobortis.',
    'Detail C: Sed eu justo vitae metus malesuada fermentum.',
    'Detail D: Fusce at velit at libero tincidunt blandit.'
]
colors = ['gold', 'yellow', 'coral', 'blue']

# Create figure and axis
fig, ax = plt.subplots()

# Plot pie chart
wedges, texts, autotexts = ax.pie(sizes, labels=products, colors=colors, autopct='%1.1f%%', startangle=140,shadow=1, wedgeprops={'edgecolor': 'black'})

# Add detail annotations
for i, text in enumerate(texts):
    text.set_text(f'{products[i]}\n{details[i]}')
    text.set_fontsize(10)
    text.set_multialignment('left')

plt.title('Product Details - Pie Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data preparation (hypothetical data)
products = ['Laptop', 'Smartphone', 'Tablet', 'Smartwatch', 'Camera', 'Headphones']
states = ['California', 'Texas', 'New York', 'Florida', 'Illinois', 'Pennsylvania']
sales_percentages = [
    [10, 20, 15, 25, 5, 25],    # Laptop sales percentages across states
    [15, 10, 20, 5, 30, 20],    # Smartphone sales percentages across states
    [5, 25, 30, 10, 20, 10],    # Tablet sales percentages across states
    [20, 15, 10, 25, 10, 20],   # Smartwatch sales percentages across states
    [25, 5, 10, 30, 15, 15],    # Camera sales percentages across states
    [10, 30, 25, 15, 5, 15]     # Headphones sales percentages across states
]

# Product summaries
product_summaries = {
    'Laptop': 'Powerful computing devices for work and entertainment.',
    'Smartphone': 'Portable communication and computing devices with advanced features.',
    'Tablet': 'Compact touchscreen devices ideal for browsing and media consumption.',
    'Smartwatch': 'Wearable devices that track health metrics and offer smartphone connectivity.',
    'Camera': 'High-quality imaging devices for capturing photos and videos.',
    'Headphones': 'Personal audio devices for immersive music and audio experiences.'
}

# Calculate total sales percentages for each product (for pie chart)
total_sales_percentages = [sum(percentages) for percentages in sales_percentages]

# Plotting the pie chart
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(total_sales_percentages, labels=products, autopct='%1.1f%%', startangle=90)

# Add product summaries as annotations
for text in texts:
    product_name = text.get_text()
    if product_name in product_summaries:
        text.set_text(f'{product_name}\n\n{product_summaries[product_name]}')

ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Sales Distribution of Products Across States')
plt.show()
```
