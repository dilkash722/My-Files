```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'class3.xlsx'  # Replace with the actual file path
df = pd.read_excel('class3.xlsx')

# Convert the relevant columns to numeric types (if they contain any non-numeric values)
numeric_columns = ['URDU', 'DINIYAAT', 'HINDI', 'ENGLISH', 'MATH']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Bar chart: Comparing obtained marks in subjects
plt.figure(figsize=(60,30))
df.set_index('ROLL NO')[numeric_columns].plot(kind='bar', stacked=True)
plt.title('Obtained Marks by Subject for Each Student')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Subjects')
plt.tight_layout()
plt.show()

# Pie chart: Distribution of students across divisions
division_counts = df['Division'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(division_counts, labels=division_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'orange'])
plt.title('Distribution of Students by Division')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
df = pd.read_excel('class3.xlsx')  # Adjust the path if necessary

# Display the first few rows to understand the structure of the data
print(df.head())

# Group by 'Division' and calculate the mean OBT marks
# Replace 'Division' and 'OBT Marks' with actual column names if different
division_obt_means = df.groupby('DIVISION')['OBT MARKS'].mean()

# Plotting the chart
plt.figure(figsize=(10, 6))
division_obt_means.plot(kind='bar', color='skyblue')
plt.title('Average OBT Marks by Division')
plt.xlabel('Division')
plt.ylabel('Average OBT Marks')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files for each class
file_class_1 = "Student_Results - CLASS 1.csv"
file_class_2 = "Student_Results - CLASS 2.csv"
file_class_3 = "Student_Results - CLASS 3.csv"
file_class_4 = "Student_Results - CLASS 4.csv"

# Reading the CSV files
class_1_data = pd.read_csv(file_class_1)
class_2_data = pd.read_csv(file_class_2)
class_3_data = pd.read_csv(file_class_3)
class_4_data = pd.read_csv(file_class_4)

# Extract division data
class_1_divisions = class_1_data['Division'].value_counts()
class_2_divisions = class_2_data['Division'].value_counts()
class_3_divisions = class_3_data['Division'].value_counts()
class_4_divisions = class_4_data['Division'].value_counts()

# Combine the division counts for each class into one dataframe for easier plotting
division_counts = pd.DataFrame({
    'Class 1': class_1_divisions,
    'Class 2': class_2_divisions,
    'Class 3': class_3_divisions,
    'Class 4': class_4_divisions
}).fillna(0)  # Fill NaN with 0 where a class doesn't have a particular division

# Plotting the pie charts for divisions in each class
fig, axes = plt.subplots(2, 2, figsize=(15, 10))  # Correcting the figure size for better layout

# Class 1 pie chart
axes[0, 0].pie(class_1_divisions, labels=class_1_divisions.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
axes[0, 0].set_title('Class 1 Divisions')

# Class 2 pie chart
axes[0, 1].pie(class_2_divisions, labels=class_2_divisions.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
axes[0, 1].set_title('Class 2 Divisions')

# Class 3 pie chart
axes[1, 0].pie(class_3_divisions, labels=class_3_divisions.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
axes[1, 0].set_title('Class 3 Divisions')

# Class 4 pie chart
axes[1, 1].pie(class_4_divisions, labels=class_4_divisions.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
axes[1, 1].set_title('Class 4 Divisions')

# Adjust the layout to prevent overlap and display the charts properly
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files for each class
file_class_1 = "Student_Results - CLASS 1.csv"
file_class_2 = "Student_Results - CLASS 2.csv"
file_class_3 = "Student_Results - CLASS 3.csv"
file_class_4 = "Student_Results - CLASS 4.csv"

# Reading the CSV files
class_1_data = pd.read_csv(file_class_1)
class_2_data = pd.read_csv(file_class_2)
class_3_data = pd.read_csv(file_class_3)
class_4_data = pd.read_csv(file_class_4)

# Extract division data
class_1_divisions = class_1_data['Division'].value_counts()
class_2_divisions = class_2_data['Division'].value_counts()
class_3_divisions = class_3_data['Division'].value_counts()
class_4_divisions = class_4_data['Division'].value_counts()

# Combine the division counts for each class into one dataframe for easier plotting
division_counts = pd.DataFrame({
    'Class 1': class_1_divisions,
    'Class 2': class_2_divisions,
    'Class 3': class_3_divisions,
    'Class 4': class_4_divisions
}).fillna(0)  # Fill NaN with 0 where a class doesn't have a particular division

# Bright and vibrant color palette
colors = ['#FF6666', '#FFCC66', '#66CCFF', '#99FF99']  # Bright colors

# Plotting the pie charts for divisions in each class
fig, axes = plt.subplots(2, 2, figsize=(15, 10))  # Create a 2x2 grid of subplots

# Class 1 pie chart
axes[0, 0].pie(class_1_divisions, labels=class_1_divisions.index, autopct='%1.1f%%', 
               colors=colors)
axes[0, 0].set_title('Class 1 Passing Percentage')

# Class 2 pie chart
axes[0, 1].pie(class_2_divisions, labels=class_2_divisions.index, autopct='%1.1f%%', 
               colors=colors)
axes[0, 1].set_title('Class 2 Passing Percentage')

# Class 3 pie chart
axes[1, 0].pie(class_3_divisions, labels=class_3_divisions.index, autopct='%1.1f%%', 
               colors=colors)
axes[1, 0].set_title('Class 3 Passing Percentage')

# Class 4 pie chart
axes[1, 1].pie(class_4_divisions, labels=class_4_divisions.index, autopct='%1.1f%%', 
               colors=colors)
axes[1, 1].set_title('Class 4 Passing Percentage')

# Adjust the layout to ensure everything fits and shows correctly
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.offline as pyo

# Load the CSV files for each class
file_class_1 = "Student_Results - CLASS 1.csv"
file_class_2 = "Student_Results - CLASS 2.csv"
file_class_3 = "Student_Results - CLASS 3.csv"
file_class_4 = "Student_Results - CLASS 4.csv"

# Reading the CSV files
class_1_data = pd.read_csv(file_class_1)
class_2_data = pd.read_csv(file_class_2)
class_3_data = pd.read_csv(file_class_3)
class_4_data = pd.read_csv(file_class_4)

# Extract division data
class_1_divisions = class_1_data['Division'].value_counts()
class_2_divisions = class_2_data['Division'].value_counts()
class_3_divisions = class_3_data['Division'].value_counts()
class_4_divisions = class_4_data['Division'].value_counts()

# Create a list of pie charts (one for each class)
fig = go.Figure()

# Add Class 1 Pie Chart
fig.add_trace(go.Pie(
    labels=class_1_divisions.index,
    values=class_1_divisions,
    name='Class 1',
    title='Class 1 Divisions',
    domain=dict(row=0, column=0)
))

# Add Class 2 Pie Chart
fig.add_trace(go.Pie(
    labels=class_2_divisions.index,
    values=class_2_divisions,
    name='Class 2',
    title='Class 2 Divisions',
    domain=dict(row=0, column=1)
))

# Add Class 3 Pie Chart
fig.add_trace(go.Pie(
    labels=class_3_divisions.index,
    values=class_3_divisions,
    name='Class 3',
    title='Class 3 Divisions',
    domain=dict(row=1, column=0)
))

# Add Class 4 Pie Chart
fig.add_trace(go.Pie(
    labels=class_4_divisions.index,
    values=class_4_divisions,
    name='Class 4',
    title='Class 4 Divisions',
    domain=dict(row=1, column=1)
))

# Update layout for subplots
fig.update_layout(
    title_text="Division Distribution Across Classes",
    grid=dict(rows=2, columns=2),  # Arranging the plots in 2x2 grid
    height=700,  # Adjust height
    width=1000  # Adjust width
)

# Save the figure to an HTML file
pyo.plot(fig, filename='division_pie_charts.html', auto_open=False)
```
