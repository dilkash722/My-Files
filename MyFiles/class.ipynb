```python
class Student:
    def __init__(self, name, roll_num, stream, subjects):
        self.name = name
        self.roll_num = roll_num
        self.stream = stream
        self.subjects = subjects

    def print_details(self):
        print(f"Name: {self.name}")
        print(f"Roll Number: {self.roll_num}")
        print(f"Stream: {self.stream}")
        for subject, marks in self.subjects.items():
            print(f"{subject}: {marks} marks")
        print("*" * 50)

class Students:
    def __init__(self):
        self.students_list = []

    def add_student(self, name, roll_num, stream, subjects):
        student = Student(name, roll_num, stream, subjects)
        self.students_list.append(student)

    def print_all_students(self):
        for student in self.students_list:
            student.print_details()

# Students class instance
students = Students()

# Adding 10 students
students.add_student("Aakash", 1, "Science", {"Math": 85, "Physics": 90, "Chemistry": 88, "Biology": 75, "English": 80})
students.add_student("Ramesh", 2, "Commerce", {"Accountancy": 80, "Business Studies": 85, "Economics": 78, "Math": 72, "English": 70})
students.add_student("Suresh", 3, "Arts", {"History": 65, "Geography": 70, "Political Science": 68, "Economics": 60, "English": 75})
students.add_student("Mahesh", 4, "Science", {"Math": 92, "Physics": 88, "Chemistry": 85, "Biology": 78, "English": 82})
students.add_student("Rajesh", 5, "Commerce", {"Accountancy": 76, "Business Studies": 80, "Economics": 72, "Math": 70, "English": 68})
students.add_student("Dinesh", 6, "Arts", {"History": 60, "Geography": 62, "Political Science": 65, "Economics": 55, "English": 70})
students.add_student("Mukesh", 7, "Science", {"Math": 89, "Physics": 85, "Chemistry": 80, "Biology": 74, "English": 78})
students.add_student("Nilesh", 8, "Commerce", {"Accountancy": 82, "Business Studies": 87, "Economics": 80, "Math": 75, "English": 73})
students.add_student("Hitesh", 9, "Arts", {"History": 70, "Geography": 75, "Political Science": 72, "Economics": 68, "English": 78})
students.add_student("Rakesh", 10, "Science", {"Math": 95, "Physics": 90, "Chemistry": 92, "Biology": 85, "English": 88})

# Print details of all students
students.print_all_students()
```
